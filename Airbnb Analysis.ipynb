```python
#dataframe with necessary columns
new_airbnb  = airbnb[['id', 'name', 'host_id', 'host_name', 'neighbourhood_group',
       'neighbourhood', 'room_type', 'price','minimum_nights' ,'number_of_reviews', 'calculated_host_listings_count',
       'availability_365']]
new_airbnb
```

```python
#Different Neighbourhood and Hosts and their count of listings


most_booked_areas = new_airbnb.groupby(["host_name" , 'neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
most_booked_areas.sort_values(by = 'calculated_host_listings_count' , ascending = False).head(5)
```

```python
#Most number of listings -> Brooklyn 
#Host Names -> Erica , Lisel , John  ,  Mark , Obed
```

```python
#Most Booked Room-Type


most_booked_roomtype= new_airbnb.groupby([ 'room_type'])[ 'minimum_nights'].max().reset_index()
most_booked_roomtype.sort_values(by = 'minimum_nights' , ascending = False).head(10)
```

```python
#Most Booked Room Type  -> Entire home/apt
          
```

```python
# creating the horizontal bar plot

fig = plt.figure(figsize = (10, 4))
room = most_booked_roomtype['room_type']
days =most_booked_roomtype['minimum_nights']
plt.barh(room , days )
 

plt.ylabel("Room-Type")
plt.xlabel("Days Spent")
plt.title("Most-Booked-Room Type")
plt.show()
```

```python
#Most Booked Areas


most_reviewed_areas = new_airbnb.groupby([ 'neighbourhood_group'])['number_of_reviews'].max().reset_index()
most_reviewed_areas.sort_values(by = 'number_of_reviews' , ascending = False).head(10)
```

```python
# creating the bar plot

fig = plt.figure(figsize = (10, 5))
area = most_reviewed_areas['neighbourhood_group']
review = most_reviewed_areas['number_of_reviews']
 

bar_labels = most_reviewed_areas['neighbourhood_group']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange' ]

plt.bar(area, review, label=bar_labels, color=bar_colors , width = 0.4)
 
plt.xlabel("area")
plt.ylabel("review")
plt.title("Areas by Reviews")
plt.show()
```

```python
#Most Reviewed Hosts

busiest_hosts = new_airbnb.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_hosts = busiest_hosts.sort_values(by='number_of_reviews', ascending=False).head(10)
busiest_hosts
```

```python
# creating the bar plot

name = busiest_hosts['host_name']
reviews = busiest_hosts['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(name, reviews, color ='maroon',
        width = 0.4)
 
plt.xlabel("Name of the Host")
plt.ylabel("Number of Reviews")
plt.title("Most Reviewed Hosts")
plt.show()
```

```python
priceVsReviews = new_airbnb.groupby([  'price' ])['number_of_reviews'].max().reset_index()
priceVsReviews.sort_values(by = 'number_of_reviews' , ascending = False).head(10)
```

```python
#price Analysis ->100-200 price got the most reviews
```

```python
# creating the scatter plot

area = priceVsReviews['price']
price = priceVsReviews['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
plt.scatter(area, price)
plt.xlabel("Price")
plt.ylabel("Number of Review")
plt.title("Price vs Number of Reviews")
plt.show()
```

```python
corr = new_airbnb.corr(method = "kendall")
fig = plt.figure(figsize = (12, 6))
sns.heatmap(corr , annot = True)
new_airbnb.columns
```
